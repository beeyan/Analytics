```
# CPU上にtensorを作成
t = torch.tensor([[1, 2], [3, 4.]])
```

```
# 100行10列tensorの作成
t = torch.zeros(100, 10).to("cuda:0")
t[:5]
```

```
# cudaにいるtensorは一度cpuにコピーしてから渡す
torch.tensor([[1, 2], [3, 4.]], device='cuda:0')
t.to('cpu').numpy().astype('int32')
```

```
t = torch.tensor([[1, 2, 3], [4, 5, 6.]])

#(0, 2)成分の抽出
t[0, 2]
```

```
# 全行の1列目まで
t[:, :2]
```

```
# 全行の1, 3列目を抽出する
t[:, [0, 2]]
```

```
# マスク配列を用いて条件抽出
t[t > 2]
```

```
# 特定の成分に値を代入する
t[0, 1] = 100
t
```

```
# 2列目に値を代入
t[:, 1] = 200
t
```

```
# マスクした条件に値を代入
t[t > 10] = 20
t
```

### 配列演算

```
x1 = torch.tensor([[1, 2], [3, 4.]])
x2 = torch.tensor([[10, 20, 30], [40, 50, 60.]])

# 4行1列
x1.view(4, 1)
```

```
#-1は残りの次元全てを表す. 
x1.view(1, -1)
```

```
# 配列の転置
x2.t()
```

```
# 列方向の追加
torch.cat([x1, x2], dim=1)
```

```
# 行方向の追加
torch.cat([x1, x2[:, :2]], dim=0)
```

```
# 別の次元の方向に追加
torch.stack([x1, x2[:, :2]])
```

### 行列演算

```
m = torch.randn(100, 10)
v = torch.randn(10)
```

```
# inner product
torch.dot(v, v)
```

```
torch.mv(m, v)
```

```
torch.mm(m.t(), m)
```

```
# 特異値分解
u, s, v = torch.svd(m)
print(u.size())
print(s.size())
print(v.size())
```

```
symm = torch.mm(m.t(), m)
# 固有値分解
torch.eig(symm)
```

### 自動微分

```
x = torch.randn(100, 3)
a = torch.tensor([1, 2, 3.], requires_grad=True)

y = torch.mv(x, a)
o = y.sum()

# 微分の実行
o.backward()
```

```
o != x.sum(0)
```
